### Frequency array

```python
import random
l=[random.randint(0,100000) for i in range(1000)]

frq=[0]*100009
mx=0
nm=0
for i in l:
    frq[i]+=1
for i in range(100001):
    if frq[i]>mx:
        mx=frq[i]
        nm=i
print(nm,mx)
```
