```python
#problem 1
string=input("Enter a string: ")
vowels="aeiouAEIOU"
count=0
for char in string:
    if char in vowels:
        count+=1
print("Number of vowels:", count)
```

```python
#problem 2
string=input("Enter a string: ")
target="iti"
print("Number of occurrences of 'iti':", string.count(target))
```

```python
#problem 3
string = input("Enter a string: ")
vowels="aeiouAEIOU"
for char in vowels:
    string = string.replace(char, "")
print("String after removing vowels:", string)
```

```python
#problem 4
string=input("Enter a string: ")
print(string)
for i in range(len(string)):
    if string[i]=='i':
        print("Index of 'i':", i)
```

```python
#problem 5
num = input("Enter a number: ")
if num.isdigit()==0:
    print("متهزرش معايا")
else:
    num = int(num)
    for i in range(1,num+1):
        s=""
        for j in range(1,i+1):
            s+=str(j*i) + ' '
        print(s)
```

```python
#problem 6
num = input("Enter a number: ")
if num.isdigit()==0:
    print("متهزرش معايا")
else:
    num = int(num)
    for i in range(1,num+1):
        s=" "*(num-i) + "*"*(i)
        print(s)
```
