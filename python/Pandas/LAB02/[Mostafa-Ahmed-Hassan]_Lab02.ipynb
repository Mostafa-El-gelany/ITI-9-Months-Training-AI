**First issue**

all numbers has , instead of .


```python
#solution 
df = df.replace(',', '.', regex=True)

```

```python
#test 
df.head()
```

**Second issue**

most of the columns are not numeric

```python
#solution 
df[df.columns[2:]]=df[df.columns[2:]].apply(pd.to_numeric)
```

```python
#test 
df.info()
```

**Third issue**

Some cells has null values

```python
#solution
df.loc[:, df.columns[2:]] = df.loc[:, df.columns[2:]].fillna(df.loc[:, df.columns[2:]].mean())
```

```python
#test
df.isnull().sum().sum()
```

**Fourth issue**

Outliers
in the area and population columns there is outliers in mathematical perspective but they are correct such as china or russia so it is not correct to transform them

```python
for i in range(2,20,1):
    Q1 = df[df.columns[i]].quantile(0.25)
    Q3 = df[df.columns[i]].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[df.columns[i]] < Q1 - 1.5*IQR) | (df[df.columns[i]] > Q3 + 1.5*IQR)]
    print(f'Number of outliers in column {df.columns[i]}: {outliers.shape[0]}')
```

```python
#print rows with outliers in any column
for i in range(2,20,1):
    Q1 = df[df.columns[i]].quantile(0.25)
    Q3 = df[df.columns[i]].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[df.columns[i]] < Q1 - 1.5*IQR) | (df[df.columns[i]] > Q3 + 1.5*IQR)]
    print(f'Outlier rows in column {df.columns[i]}:')
    print(outliers)
```

```python
#print rows with outliers in any column
Q1 = df[df.columns[2]].quantile(0.25)
Q3 = df[df.columns[2]].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df[df.columns[2]] < Q1 - 1.5*IQR) | (df[df.columns[2]] > Q3 + 1.5*IQR)]
print(f'Outlier rows in column {df.columns[2]}:')
print(outliers)
```

```python
for i in range(4,20,1):
    Q1 = df[df.columns[i]].quantile(0.25)
    Q3 = df[df.columns[i]].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    df.loc[df[df.columns[i]] < lower_bound, df.columns[i]] = lower_bound
    df.loc[df[df.columns[i]] > upper_bound, df.columns[i]] = upper_bound
```

```python
#test outliers 
for i in range(2,20,1):
    Q1 = df[df.columns[i]].quantile(0.25)
    Q3 = df[df.columns[i]].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[df.columns[i]] < Q1 - 1.5*IQR) | (df[df.columns[i]] > Q3 + 1.5*IQR)]
    print(f'Number of outliers in column {df.columns[i]} after transformation: {outliers.shape[0]}')
```
