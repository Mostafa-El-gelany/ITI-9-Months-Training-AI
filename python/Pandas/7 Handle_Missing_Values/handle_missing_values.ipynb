```python
filling_values = {"weight": 0, "hindfoot_length": df['hindfoot_length'].mean(), "sex": df['sex'].mode().iloc[0], "species_id": df['species_id'].mode().iloc[0]}
df_5 = df.fillna(value=filling_values)
df_5
```
