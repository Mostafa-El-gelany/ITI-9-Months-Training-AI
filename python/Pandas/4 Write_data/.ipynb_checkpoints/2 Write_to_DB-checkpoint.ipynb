```python
import pandas as pd
#!pip install psycopg2
import psycopg2
```

```python
# Connection parameters, yours will be different
params_dic = {
    "host"      : "localhost",
    "database"  : "postgres",
    "user"      : "postgres",
    "password"  : "postgres"
}

conn = psycopg2.connect(**params_dic)
```

```python
new_shippers_df = pd.DataFrame({ 
                         'shipper_id' : [50, 51, 52],
                         'company_name' : ["ITI_50", "Edafa_50", "Nokia_50"],
                         'phone' : ["010", "011", "012"]
                      })
new_shippers_df
```

```python
for ind in new_shippers_df.index:
    query = """INSERT into shippers(shipper_id, company_name, phone) values(%d,'%s','%s');""" % (new_shippers_df['shipper_id'][ind], new_shippers_df['company_name'][ind], new_shippers_df['phone'][ind])
    print(query)
    cursor = conn.cursor()
    cursor.execute(query)
    conn.commit()
```

```python
conn.close()
```
