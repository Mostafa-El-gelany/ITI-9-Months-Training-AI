### Grouping
one or more of the following steps:
- Splitting the data into groups based on some criteria
- Applying a function to each group independently
- Combining the results into a data structure

```python
import pandas as pd
```

```python
employees_df = pd.read_excel('../datasets/EmployeesWithGrades.xlsx', 'Sheet1')
```

```python
employees_df
```

### Group by Department

calculate total years of service by employees in each department

```python
employees_df['Department'].value_counts()
```

```python
employees_df.groupby('Department')
```

```python
employees_df.groupby('Department').mean()
```

```python
employees_df.groupby('Department').mean()['Age'].reset_index()
```

```python
employees_df.groupby('Grade').count()
```

```python
employees_df.groupby('Grade')['Age'].max()
```

```python
employees_df.groupby('Grade').max()['Age']
```
