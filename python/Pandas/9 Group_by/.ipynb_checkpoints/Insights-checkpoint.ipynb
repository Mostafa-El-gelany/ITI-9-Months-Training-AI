```python
import pandas as pd
```

```python
data = {
    "OrderID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Product": ["Laptop", "Laptop", "Phone", "Phone", "Tablet", "Tablet", "Laptop", "Phone"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics"],
    "Quantity": [1, 2, 1, 3, 2, 1, 1, 2],
    "Price": [1000, 1200, 800, 750, 400, 450, 1100, 850],
    "Total_Sales": [1000, 2400, 800, 2250, 800, 450, 1100, 1700],
    "Date": pd.to_datetime(["2024-02-01", "2024-02-01", "2024-02-02", "2024-02-03", "2024-02-04", "2024-02-04", "2024-02-05", "2024-02-05"]),
    "Region": ["Cairo", "Alex", "Cairo", "Alex", "Cairo", "Cairo", "Alex", "Alex"]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df["year"]=df["Date"].dt.year
```

```python
df.info()
```

```python
df.head()
```

# Total Sales by Product

```python
df_grouped = df.groupby("Product")["Total_Sales"].sum().reset_index()
print(df_grouped)
# Identify which product generates the highest total sales.
```

 # Number of Orders per Product

```python
df_grouped = df.groupby("Product")["OrderID"].count().reset_index().rename(columns={"OrderID":"Order Count"})
print(df_grouped)
#Find out the most frequently ordered products
```

 # Total Sales by Region

```python
df_grouped = df.groupby("Region")["Total_Sales"].sum().reset_index()
df_grouped
#Identify the most profitable regions.
```

#  Total Sales by Date

```python
df_grouped = df.groupby(df["Date"])["Total_Sales"].sum().reset_index()
df_grouped
#Analyze daily sales trends/
```

 # Total Sales by Product and Region

```python
df_grouped = df.groupby(["Product", "Region"])["Total_Sales"].sum().reset_index()
df_grouped
```

# Total Quantity Sold per Product

```python
df_grouped = df.groupby("Product")["Quantity"].sum().reset_index()
print(df_grouped)
#Identify which products have the highest number of units sold
```

```python
data = {
    "OrderID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Product": ["Laptop", "Laptop", "Phone", "Phone", "Tablet", "Tablet", "Laptop", "Phone"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics"],
    "Quantity": [1, 2, 1, 3, 2, 1, 1, 2],
    "Price": [1000, 1200, 800, 750, 400, 450, 1100, 850],
    "Total_Sales": [1000, 2400, 800, 2250, 800, 450, 1100, 1700],
    "Date": pd.to_datetime(["2024-02-01", "2024-02-01", "2024-02-02", "2024-02-03", "2024-02-04", "2024-02-04", "2024-02-05", "2024-02-05"]),
    "Region": ["Cairo", "Alex", "Cairo", "Alex", "Cairo", "Cairo", "Alex", "Alex"]
}
```

```python
df = pd.DataFrame(data)
df
```

```python

df['Product'].fillna('Unknown', inplace=True)


df['Price'].fillna(df['Price'].mean(), inplace=True)


df['Quantity'] = df['Quantity'].apply(lambda x: max(x, 0))

print(df)
```
