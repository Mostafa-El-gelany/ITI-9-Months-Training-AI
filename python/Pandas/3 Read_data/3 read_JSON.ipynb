```python
df = pd.concat([df, pd.json_normalize(df['temp'])], axis=1)
df = df.rename(columns={'day': 'temp_day', 'min': 'temp_min', 'max': 'temp_max',
                               'night': 'temp_night', 'eve': 'temp_eve', 'morn': 'temp_morn'})
df.head(2)
```

```python
df = pd.concat([df, pd.json_normalize(df['feels_like'])], axis=1)
```

### We can get the JSON dataset as API response, such as the following:

```python
import requests

url = "https://community-open-weather-map.p.rapidapi.com/forecast/daily"
querystring = {"q":"Cairo"}
headers = {
    'X-RapidAPI-Host': "community-open-weather-map.p.rapidapi.com",
    'X-RapidAPI-Key': "" #Kindly put your API-Key here :)
    }

response = requests.request("GET", url, headers=headers, params=querystring)
print(response.text)
```

```python
response
```

```python
type(response.text)
```

```python
json.loads(response.text)
#take care that it's json.loads not json.load
```

```python
df_from_API = pd.DataFrame(json.loads(response.text)['list'])
df_from_API.head(2)
```
