```python
df['receiverDailyTransactionSum']=df.groupby(['nameDest','day'])['amount'].transform('sum')
```

```python
df['transactionScore']=df[['senderGapScore','receiverGapScore','amountScore','senderDailyTransactionSumScore','senderDailyTransactionCountScore','receiverDailyTransactionSumScore','receiverDailyTransactionCountScore']].max(axis=1)
```

```python
df[df.transactionScore>9]
```

```python
df['senderScore'] = df.groupby('nameOrig')[['senderGapScore','senderDailyTransactionSumScore','senderDailyTransactionCountScore','amountScore']].transform('max').max(axis=1)
```

```python
df['receiverScore'] = df.groupby('nameDest')[['receiverGapScore','receiverDailyTransactionSumScore','receiverDailyTransactionCountScore','amountScore']].transform('max').max(axis=1)
```

```python
df[(df.senderGapScore>3) | (df.receiverGapScore>3)]
```

```python
df['senderMaxAmount'] = df.groupby('nameOrig')['amount'].transform('max')
df['receiverMaxAmount'] = df.groupby('nameDest')['amount'].transform('max')
df['senderAverageAmount'] = df.groupby('nameOrig')['amount'].transform('mean')
df['receiverAverageAmount'] = df.groupby('nameDest')['amount'].transform('mean')
df['senderTotalAmount'] = df.groupby('nameOrig')['amount'].transform('sum')
df['receiverTotalAmount'] = df.groupby('nameDest')['amount'].transform('sum')
```

```python
df['senderTransactionsCount'] = df.groupby('nameOrig')['amount'].transform('count')
df['receiverTransactionsCount'] = df.groupby('nameDest')['amount'].transform('count')
```

```python
# 1. Start with the default
df['transactionRisk'] = 'Low'

# 2. Update based on thresholds
df.loc[df['transactionScore'].abs() > 1, 'transactionRisk'] = 'Medium'
df.loc[df['transactionScore'].abs() > 3, 'transactionRisk'] = 'High'
df.loc[df['transactionScore'].abs() > 9, 'transactionRisk'] = 'Critical'
```

```python
df['senderRisk']='Low'
df.loc[df['senderScore'].abs() > 1, 'senderRisk'] = 'Medium'
df.loc[df['senderScore'].abs() > 3, 'senderRisk'] = 'High'
df.loc[df['senderScore'].abs() > 9, 'senderRisk'] = 'Critical'
```

```python
df['receiverRisk']='Low'
df.loc[df['receiverScore'].abs() > 1, 'receiverRisk'] = 'Medium'
df.loc[df['receiverScore'].abs() > 3, 'receiverRisk'] = 'High'
df.loc[df['receiverScore'].abs() > 9, 'receiverRisk'] = 'Critical'
```
